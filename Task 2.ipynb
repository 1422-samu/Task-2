```python
df.columns=["State","Date","Frequency","Estimated Unemployment Rate","Estimated Employed","Estimated Labour Participation Rate","Region","longitude","latitude"]
```

```python
plt.figure(figsize=(10, 10))
plt.bar(df["Region"], df["Estimated Unemployment Rate"])
plt.xticks(rotation=90)
```

# Interpretation : South region has the highest and West region has the lowest  Estimated Unemployment Rate

```python
plt.figure(figsize=(12, 10));
sns.catplot(x="State", y="Estimated Unemployment Rate", data=df, kind="bar");
plt.xticks(rotation=90);
```

# Interpretation : Among all the states Haryana has the highest and Meghalaya has the lowest Estimated Unemployment Rate

```python
plt.figure(figsize=(12,10))
plt.title("India's Unemployment")
sns.histplot(x="Estimated Unemployment Rate",hue="Region",data=df)
plt.show()
```
